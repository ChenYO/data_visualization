```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
squares = [1, 4, 9, 16, 25]
plt.plot(squares)
plt.show()
```

# 簡單的繪製

```python
input_value = [1, 2, 3, 4, 5]
plt.plot(input_value, squares, linewidth=5)
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)
plt.tick_params(axis="both", labelsize=14)
plt.show()
```

# Scatter

```python
plt.scatter(2, 4)
plt.show()
```

```python
plt.scatter(2, 4, s=200)
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Number", fontsize=14)
plt.tick_params(axis="both", which="major", labelsize=14)
plt.show()
```

```python
x_value = [1, 2, 3, 4, 5]
y_value = [1, 4, 9, 16, 25]
plt.scatter(x_value, y_value, s=100)
plt.show()
```

# 自動計算資料

```python
x_value = list(range(1, 1001))
y_value = [x**2 for x in x_value]
plt.scatter(x_value, y_value, s=40)
plt.axis([0, 1100, 0, 1100000])
plt.show()
```

```python
x_value = list(range(1, 1001))
y_value = [x**2 for x in x_value]
plt.scatter(x_value, y_value, c=(0, 0, 0.8), edgecolors="none", s=40)
plt.axis([0, 1100, 0, 1100000])
plt.show()
```

```python
x_value = list(range(1, 1001))
y_value = [x**2 for x in x_value]
plt.scatter(x_value, y_value, c=y_value, cmap=plt.cm.Blues, edgecolors="none", s=40)
plt.axis([0, 1100, 0, 1100000])
plt.show()
```
